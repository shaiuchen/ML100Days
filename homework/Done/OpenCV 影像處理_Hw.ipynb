```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena.png')
```

### 第一大題 (OpenCV 標註訓練)

1. 將 Lena 的臉蛋用紅框框起來
2. 畫出紅匡中的對角線
3. 標注紅匡左上角的座標
4. 利用 subplot 將三張圖片繪製出來

```python

# 獲得用紅匡標注的圖片
img_rect = img.copy()
cv2.rectangle(img_rect, (200, 150), (390, 420), (0, 0, 255), 3)

# 獲得用對角線標注的圖片
img_line = img.copy()
cv2.line(img_line, (200, 150), (390, 420), (0, 0, 255), 3)

# 獲得具有座標標注的圖片
img_text = img.copy()
cv2.putText(img_text, '(200, 150)', (200, 150), 0, 1, (0, 0, 255), 2)

# 利用 subplot 將三張圖片繪製出來
display_imgs = [img_rect, img_line, img_text]
display_titles = ['Rectangle', 'Line', 'Text']

plt.figure(figsize = (15, 5))
for i in range(3):
    plt.subplot(1, 3, i + 1, title = display_titles[i])
    plt.imshow(display_imgs[i][:, :, [2, 1, 0]])
    plt.axis('off')

plt.show()
```

### 第二大題 (繪製 Histogram)

1. 獲得 3 個 channel (RGB) 的 histogram
2. 利用 subplot 繪製出來

```python
plt.figure(figsize = (8, 14))
for i in range(3):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.subplot(3, 1, i + 1, title = 'channel {} histogram'.format(i))
    plt.bar(range(0, 256), hist[:, 0])
plt.show()
```
